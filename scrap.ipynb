```python


def convert_format1_date(date_str):
    try:
       
        start, end_year = date_str.split(' - ')
        start_date = pd.to_datetime(start).strftime('%d/%m/%Y')
        end_date = pd.to_datetime(end_year).strftime('%d/%m/%Y')
        return f"{start_date} - {end_date}"
    except ValueError:
        return date_str




df1['date'] = df1['date'].apply(convert_format1_date)




def extract_dates(date_str):
    try:
        dates = date_str.split(' - ')
        
        start_date_parts = dates[0].split(' ')
        start_day = start_date_parts[1]
        start_month = start_date_parts[0]
        start_year = dates[1].split(', ')[1] 

        start_date = "{:02d}/{:02d}/{}".format(int(start_day), 
                                                (pd.to_datetime(start_month, format='%b').month),
                                                int(start_year))
        
        end_date_parts = dates[1].split(' ')
        end_day = end_date_parts[1].replace(',', '') 
        end_month = end_date_parts[0]
        
      
        end_date = "{:02d}/{:02d}/{}".format(int(end_day), 
                                              (pd.to_datetime(end_month, format='%b').month),
                                              int(start_year))
        
        combined_date = f"{start_date} - {end_date}"
        
        return combined_date
    except:
        return date_str


df1['date'] = df1['date'].apply(extract_dates)




def extract_dates(date_str):
    try:
        if '-' in date_str and ',' in date_str:
        
            dates = date_str.split(' - ')

       
            start_date_parts = dates[0].split(' ')
            start_day = start_date_parts[1]
            start_month = start_date_parts[0]
            start_year = dates[1].split(', ')[1]  

 
            start_date = "{:02d}/{:02d}/{}".format(int(start_day),
                                                    (pd.to_datetime(start_month, format='%b').month),
                                                    int(start_year))

      
            end_date_parts = dates[1].split(', ')[0].split(' ')
            if len(end_date_parts) == 2:
                end_day = end_date_parts[1]
            else:
                end_day = dates[1].split(', ')[0]
            end_month = start_month  
            end_year = start_year 

            end_date = "{:02d}/{:02d}/{}".format(int(end_day),
                                                  (pd.to_datetime(end_month, format='%b').month),
                                                  int(end_year))
            return f"{start_date} - {end_date}"
        
        elif ',' in date_str:
            single_date_parts = date_str.split(', ')
            day_month = single_date_parts[0].split(' ')[1]
            year = single_date_parts[1]
            return "{:02d}/{:02d}/{} - {:02d}/{:02d}/{}".format(int(day_month),
                                                                 (pd.to_datetime(single_date_parts[0].split(' ')[0], format='%b').month),
                                                                 int(year),
                                                                 int(day_month),
                                                                 (pd.to_datetime(single_date_parts[0].split(' ')[0], format='%b').month),
                                                                 int(year))
        else:
            return date_str
    except:
        return date_str


df1['date'] = df1['date'].apply(lambda x: extract_dates(x))

def split_dates(date_str):
    start, end = date_str.split(' - ')
    return start, end

df1[['Start date', 'End date']] = df1['date'].apply(lambda x: pd.Series(split_dates(x)))

df1 = df1.drop(columns=['date'])
 

```
